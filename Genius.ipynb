```python
#Import module and config with token
import requests
import pandas as pd
from pprint import pprint
import lyricsgenius
from config import genius_token
genius = lyricsgenius.Genius(genius_token)
```

```python
#Open top_love_songs csv file
top_song_df=pd.read_csv("csv/top_love_songs.csv")

#add new columns to populate with year and lyrics
top_song_df["Year"]=""
top_song_df["Lyrics"]=""
top_song_df
```

```python
# #check song object attributes obtained through lyricsgenius
# dir(song)
```

```python
#Search year and lyrics for TOP100 songs:

for index, row in top_song_df.iloc[0:100].iterrows():
    title=row["Song Name"]
    artist=row["Artist"]
    print(f'Retrieving info for Index {index}: {title} by {artist}')
    
    try:       
        song = genius.search_song(title, artist=artist)
        top_song_df.loc[index, "Year"]= song.year
        top_song_df.loc[index, "Lyrics"]= song.lyrics
        
    except:
        print(f'----- missing info for {title} by {artist}')

print("FIN")
```

```python
songs_df=top_song_df.drop("Unnamed: 0", axis=1)
songs_df
```

```python
#Drop weird column created for don't know what reason
songs_df=top_song_df.drop("Unnamed: 0", axis=1)

#Export to csv
songs_df.to_csv("csv/top100_year_lyrics.csv")
```
